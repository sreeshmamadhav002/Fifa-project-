## 5. plot a scatter plot on over all and potential

## 6.plot a histogram for salary range of players

```python
df['Wage']=df['Wage'].str.replace('€','').str.replace('K','000').astype(int)
df['Wage']
```

```python
plt.hist(df['Wage'],bins=10)
plt.xlabel('Wage')
plt.ylabel('no.of players')
plt.title('salary Distribution')
plt.show()
```
